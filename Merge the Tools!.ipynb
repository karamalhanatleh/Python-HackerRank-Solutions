```python
import textwrap

def merge_the_tools(string, k):
    # Split the string into parts of size k
    split_words = textwrap.wrap(string, width=k)

    for word in split_words:
        unique = ""
        for char in word:
            if char not in unique:
                unique += char
        print(unique)
if __name__ == '__main__':
    string, k = input(), int(input())
    merge_the_tools(string, k)
```
